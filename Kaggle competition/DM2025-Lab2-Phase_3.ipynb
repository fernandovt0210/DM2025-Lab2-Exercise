```python
import pandas as pd
import json

with open('./final_posts.json', 'r', encoding='utf-8') as json_file:
    json_data = json.load(json_file)


json_raw_DF = pd.json_normalize(json_data)

json_raw_DF
```

```python
emotions_DF_unfiltered = json_raw_DF[["root._source.post.post_id","root._source.post.text", "root._source.post.hashtags"]]

emotions_DF_unfiltered.columns = ["Post_ID","Text","Hashtags"]

emotions_DF_unfiltered
```

```python


data_identification_csv_DF = pd.read_csv("./data_identification.csv", sep= ",", names=["Post_ID_data_ID_csv","Split"], skiprows= 1)

data_identification_csv_DF
```

```python
print(data_identification_csv_DF.query("Split == 'test' ")) # 16281 rows // 25.37 %
print(data_identification_csv_DF.query("Split == 'train' ")) # 47890 rows // 74.63 %
```

```python
emotions_DF = pd.concat([emotions_DF_unfiltered, data_identification_csv_DF["Split"]], axis = 1)
emotions_DF
```

```python
def Null_counter(row):
    counts = 0
    for x in row:
        if x == True:
            counts += 1
    return counts

emotions_DF.isnull().apply(lambda x: Null_counter(x)) #No empty values
```

```python
sum(emotions_DF.duplicated('Post_ID')) #No duplicates
```

```python
emotions_csv_DF = pd.read_csv("./emotion.csv") 
emotions_csv_DF
```

```python
emotions_csv_DF.isnull().apply(lambda x: Null_counter(x)) #No empty values
```
