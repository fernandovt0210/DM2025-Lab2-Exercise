### 1.3 Explanation of Your Model

For the model I decided to use the RoBERTa version of BERT. ## 1. Preprocessing Steps

```python
##############################################################################################################################################################################################
#           
#  Important note: The code below is a "clean" version of the code that I made. Taking all of the relevant parts of it and organizing them. The "messy" code you can find it on a jupyter
#  notebook under the folder ""Kaggle competition". This is the same code, just organized. 
#
##############################################################################################################################################################################################

import pandas as pd
import json
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

with open('./Kaggle competition/final_posts.json', 'r', encoding='utf-8') as json_file:
    json_data = json.load(json_file)


json_raw_DF = pd.json_normalize(json_data)

emotions_DF_unfiltered = json_raw_DF[["root._source.post.post_id","root._source.post.text", "root._source.post.hashtags"]]

emotions_DF_unfiltered.columns = ["Post_ID","Text","Hashtags"]

data_identification_csv_DF = pd.read_csv("./Kaggle competition/data_identification.csv", sep= ",", names=["Post_ID_data_ID_csv","Split"], skiprows= 1)

emotions_DF = pd.concat([emotions_DF_unfiltered, data_identification_csv_DF["Split"]], axis = 1)

def Null_counter(row):
    counts = 0
    for x in row:
        if x == True:
            counts += 1
    return counts

emotions_DF.isnull().apply(lambda x: Null_counter(x)) #No empty values
sum(emotions_DF.duplicated('Post_ID')) #No duplicates

emotions_csv_DF = pd.read_csv("./Kaggle competition/emotion.csv") 

emotions_DF_merged = emotions_DF.merge(emotions_csv_DF, left_on= "Post_ID", right_on="id", how="left")

fig, ax = plt.subplots()
ax.pie(emotions_DF_merged.emotion.value_counts(), labels=emotions_DF_merged.emotion.value_counts().index, autopct='%1.1f%%',
       pctdistance=1.25, labeldistance=.6)

train_DF = emotions_DF_merged[emotions_DF_merged["Split"] == "train"]
joy_DF = train_DF[train_DF["emotion"] == "joy"]

joy_DF_sample = joy_DF.sample(n=2100, random_state=8)

no_joy_DF = train_DF[train_DF["emotion"] != "joy"]
no_joy_DF_sample = no_joy_DF.sample(n=4400, random_state=8)

sample_DF_train = pd.concat([joy_DF_sample, no_joy_DF_sample])

fig, ax = plt.subplots()
ax.pie(sample_DF_train.emotion.value_counts(), labels=sample_DF_train.emotion.value_counts().index, autopct='%1.1f%%',
       pctdistance=1.25, labeldistance=.6)


train_text_split , test_text_split, train_emotion_split, test_emotion_split = train_test_split(sample_DF_train["Text"], sample_DF_train["emotion"],  test_size=0.3 , random_state=8, stratify=sample_DF_train["emotion"])

RB_train, RB_test = train_test_split(sample_DF_train, test_size=0.4, random_state=8, stratify=sample_DF_train["emotion"]) # Increased the test size. Felt it necessary after the poor results of TF

```

## 3. Model Implementation Steps

```python

from datasets import Dataset
from transformers import RobertaTokenizerFast
from transformers import RobertaForSequenceClassification
from transformers import TrainingArguments
from transformers import Trainer
import torch





tokenizer = RobertaTokenizerFast.from_pretrained("roberta-base")

def tokens(batch):
    return tokenizer(
        batch["Text"],
        truncation=True,
        padding="max_length", 
        max_length=256
    )

train_DS = Dataset.from_pandas(RB_train)
test_DS = Dataset.from_pandas(RB_test)

train_DS = train_DS.map(tokens, batched= True)
test_DS = test_DS.map(tokens, batched= True)

label_2_ID = {label: i for i, label in enumerate(sample_DF_train["emotion"].unique())}
ID_2_label = {i: label for label, i in label_2_ID.items()}

train_DS = train_DS.map(lambda x: {"labels": label_2_ID[x["emotion"]]})
test_DS = test_DS.map(lambda x: {"labels": label_2_ID[x["emotion"]]})

RB_model = RobertaForSequenceClassification.from_pretrained("roberta-base", num_labels=len(label_2_ID))

training_arguments = TrainingArguments(output_dir = "./Kaggle competition/results", eval_strategy = "epoch", per_device_train_batch_size = 32, per_device_eval_batch_size = 32, num_train_epochs = 1, weight_decay = 0.01)

trainer = Trainer(model=RB_model, args=training_arguments, train_dataset=train_DS, eval_dataset=test_DS, tokenizer=tokenizer)

results = trainer.evaluate()

testing_DF_After_RB = emotions_DF_merged[emotions_DF_merged["Split"] == 'test']
texts_list = testing_DF_After_RB["Text"].tolist()


DF_for_emotion_labes = testing_DF_After_RB.copy()

texts = DF_for_emotion_labes["Text"].tolist()
post_ids = DF_for_emotion_labes["Post_ID"].tolist()

encodings = tokenizer(texts, padding=True, truncation=True, max_length=256, return_tensors="pt")

input_ids = encodings["input_ids"]
attention_mask = encodings["attention_mask"]

batch_size_for_eva = 32
predicted_label_ids = []

RB_model.eval()
with torch.no_grad():
    for i in range(0, len(input_ids), batch_size_for_eva):
        batch_input_ids = input_ids[i:i+batch_size_for_eva]
        batch_attention = attention_mask[i:i+batch_size_for_eva]

        outputs = RB_model(
            input_ids=batch_input_ids,
            attention_mask=batch_attention
        )

        batch_logits = outputs.logits
        batch_preds = torch.argmax(batch_logits, dim=1).tolist()

        predicted_label_ids.extend(batch_preds)

id2label = {
    0: "joy",
    1: "sadness",
    2: "surprise",
    3: "anger",
    4: "fear",
    5: "disgust"
}

predicted_emotions = [id2label[i] for i in predicted_label_ids]

RB_DF = pd.DataFrame({
    "id": post_ids,
    "emotion": predicted_emotions
})

RB_DF.to_csv("RB_predictions.csv", index=False)
print("Saved predictions to RB_predictions.csv")
```
